```python
import os
import shutil
import cv2
```

```python
train_img_path="/home/host_data/radio_signal_data/train/"
valid_img_path="/home/host_data/radio_signal_data/valid/"
test_img_path="/home/host_data/radio_signal_data/test/"

radio_train_img_list=os.listdir(train_img_path)
radio_valid_img_list=os.listdir(valid_img_path)
radio_test_img_list=os.listdir(test_img_path)
```

```python
# 이미지 이름 바꾸기(_jpg.rf. 없애주기)
# 이미지 옮기기
train_img_organ_path="/home/host_data/radio_signal_data_organized/train/image/"
valid_img_organ_path="/home/host_data/radio_signal_data_organized/valid/image/"
test_img_organ_path="/home/host_data/radio_signal_data_organized/test/image/"

radio_train_img_list=os.listdir(train_img_organ_path)
radio_valid_img_list=os.listdir(valid_img_organ_path)
radio_test_img_list=os.listdir(test_img_organ_path)

for file in radio_train_img_list:
#     print(file)
    if(file.endswith(".jpg")):
#         print(file.replace("jpg.rf.",""))
        shutil.move(train_img_organ_path+file,'/home/host_data/radio_signal_data_organized/train/image/'+file.replace("jpg.rf.",""))
        
for file in radio_valid_img_list:
    if(file.endswith(".jpg")):\
        shutil.move(valid_img_organ_path+file,'/home/host_data/radio_signal_data_organized/valid/image/'+file.replace("jpg.rf.",""))
        
for file in radio_test_img_list:
    if(file.endswith(".jpg")):\
        shutil.move(test_img_organ_path+file,'/home/host_data/radio_signal_data_organized/test/image/'+file.replace("jpg.rf.",""))        
```

```python
def normalize(ori_list,img_shape):
    x_cen=float(ori_list[1])/img_shape[0]
    y_cen=float(ori_list[2])/img_shape[1]
    w=float(ori_list[3])/img_shape[0]
    h=float(ori_list[4])/img_shape[1]
    norm_list=[ori_list[0], x_cen, y_cen, w, h]
    return norm_list
```

```python
# annotation 파일 개별화-train
f=open(train_img_path+"_annotations.txt",'r')
f_lines=f.readlines()
cnt=0
for line in f_lines:
    cnt=cnt+1
    text=line.split(" ")
    filename=text[0].replace("jpg.rf.","")
    img=cv2.imread(train_img_path+text[0])
    print(img.shape)
    textname=filename.split(".jpg")[0]+".txt"
    print(f"{cnt},{textname}")
    coor_list=text[1].replace("\n"," ").split(",")
    class_n_coor_list=[coor_list[4].replace(" ",""),coor_list[0],coor_list[1],coor_list[2],coor_list[3]]
    norm_list=normalize(class_n_coor_list,img.shape)
    write_str=f"{norm_list[0]} {norm_list[1]} {norm_list[2]} {norm_list[3]} {norm_list[4]}"
    new_txt_file=open("/home/host_data/radio_signal_data_organized/train/label/"+textname,'w')
    new_txt_file.write(write_str)
    
    print(f"filename:{textname} "+write_str+'\n')
```
