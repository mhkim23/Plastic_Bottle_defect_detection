```python
import os
import shutil
```

```python
neck_path="/home/host_data/PET_data/Neck/label/"
neck_label_list=os.listdir("/home/host_data/PET_data/Neck/label/")
neck_label_list
```

```python
count=0
label_src="/home/host_data/PET_data/Neck/label/"
label_des="/home/host_data/PET_data/Neck/selected_label/"

img_src="/home/host_data/PET_data/Neck/image/"
img_des="/home/host_data/PET_data/Neck/selected_image/"
for label_file in neck_label_list:
    img_file=label_file.split(".txt")[0]+'.jpg'
    neck_label_path=neck_path+label_file
    
    with open(neck_label_path, 'r') as file:
        labels = file.readlines()
    if(labels!=[]):
        count=count+1
        img_source=img_src+img_file
        img_destin=img_des+img_file
        
        label_source=label_src+label_file
        label_destin=label_des+label_file
        
        shutil.copyfile(img_source, img_destin)
        shutil.copyfile(label_source, label_destin)
        print(labels)
print(f"moved counts:{count}")
```

```python
TRAIN_NUM=210
VALID_NUM=36
TEST_NUM=25
```

```python
sel_neck_path="/home/host_data/PET_data/Neck/selected_label/"
neck_label_list=os.listdir(sel_neck_path)
len(neck_label_list)
```

```python
train_lab_list=[]
valid_lab_list=[]
test_lab_list=[]

train_count=0
val_count=0
test_count=0

for sel_lab_file in neck_label_list:
    if(train_count<TRAIN_NUM):
        train_lab_list.append(sel_lab_file)
    train_count=train_count+1

for sel_lab_file in neck_label_list:
    if(val_count<VALID_NUM and sel_lab_file not in train_lab_list):
        valid_lab_list.append(sel_lab_file)
        val_count=val_count+1

for sel_lab_file in neck_label_list:
    if(test_count<TEST_NUM and sel_lab_file not in train_lab_list and sel_lab_file not in valid_lab_list):
        test_lab_list.append(sel_lab_file)
        test_count=test_count+1


print(len(train_lab_list))
print(len(valid_lab_list))
print(len(test_lab_list))
```

```python
#train 옮기기
sel_neck_path="/home/host_data/PET_data/Neck/selected_label/"

label_src="/home/host_data/PET_data/Neck/selected_label/"
label_des="/home/host_data/PET_data/Neck/train/label/"

img_src="/home/host_data/PET_data/Neck/selected_image/"
img_des="/home/host_data/PET_data/Neck/train/image/"
for label_file in train_lab_list:
    img_file=label_file.split(".txt")[0]+'.jpg'
    
    img_source=img_src+img_file
    img_destin=img_des+img_file

    label_source=label_src+label_file
    label_destin=label_des+label_file

    shutil.copyfile(img_source, img_destin)
    shutil.copyfile(label_source, label_destin)

```
